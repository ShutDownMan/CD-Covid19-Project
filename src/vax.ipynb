## Lendo csv


```python

ddf = dd.read_csv(f'{data_path}/VAX_PR.csv',
    sep=';',
    dtype={
        "paciente_endereco_cep": str,
        "paciente_endereco_coIbgeMunicipio": str,
        "paciente_endereco_coPais": str,
        "vacina_categoria_codigo": str,
        "paciente_idade": str,
        "paciente_racaCor_codigo": str
    })

# ddf.set_index('vacina_dataAplicacao')

ddf.head()
```

```python

# remover pessoas que tomaram mais de uma dose
# ddf = ddf.drop_duplicates(subset=['paciente_id'], keep='first')

# filtrando apenas pelas primeiras doses
# ddf = ddf[ddf["vacina_descricao_dose"] == "1ª Dose"]

ddf_stripped = ddf[['vacina_dataAplicacao', 'paciente_endereco_coIbgeMunicipio']]
ddf_count = ddf_stripped.groupby(['vacina_dataAplicacao', 'paciente_endereco_coIbgeMunicipio']).agg(
    {'paciente_endereco_coIbgeMunicipio': 'count'}
)

# new_columns = ['date', 'city_code', 'vaccinated']
# ddf_count.index.rename('city_code')
ddf_count = ddf_count.rename(columns={0: 'date', 1:'code', 2:'vaccinated'})

# ddf_count.head()
```

```python

print('saving ddf count')
ddf_count.to_csv(f'{data_path}/vax_count.csv', single_file=True, index=False)
```

```python

ddf_stripped = ddf[['paciente_endereco_coIbgeMunicipio']]
ddf_snap = ddf_stripped.groupby(['paciente_endereco_coIbgeMunicipio']).agg(
    {'paciente_endereco_coIbgeMunicipio': 'count'}
)

# ddf_stripped = ddf_stripped.categorize(columns=['vacina_descricao_dose'])
# ddf_stripped['vacina_descricao_dose'] = ddf['vacina_descricao_dose'].astype('category')

# ddf_snap = dd.pivot_table(ddf_stripped,
#     index='paciente_endereco_coIbgeMunicipio',
#     columns='vacina_descricao_dose',
#     aggfunc='count')

ddf_snap = ddf_snap.rename(columns={0:'code', 1:'vaccinated'})

ddf_snap.head()
```

```python

print('saving ddf count')
ddf_snap.to_csv(f'{data_path}/vax_snap.csv', single_file=True)
```
