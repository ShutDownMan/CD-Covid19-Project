```python
import pandas as pd

dataset_size = 'full'

all_data_path = './../data/join/'
data_path = f'./../data/join/{dataset_size}/'

vax_all_data_path = './../data/vax/'
vax_data_path = f'./../data/vax/{dataset_size}/'

rem_all_data_path = './../data/rem/'
rem_data_path = f'./../data/rem/{dataset_size}/'
```

```python
# read removed

rem_ddf = pd.read_csv(f'{rem_data_path}/removed_parana.csv',
    sep=',',
    dtype={
        'date': str,
        'code': float
    })

rem_ddf.head()
```

```python
# read vaccinated

vax_ddf = pd.read_csv(f'{vax_data_path}/vax_snap.csv',
    sep=',',
    dtype={
        'code': str,
        'vaccinated': str
    }
)

# vax_ddf = vax_ddf.astype({'code': 'float64'})
vax_ddf['code'] = pd.to_numeric(vax_ddf['code'], errors='coerce')

vax_ddf.head()
```

```python
# read joined dataset

join_df =  pd.read_csv(f'{data_path}/covid_parana.csv',
    sep=',',
    dtype={},
    index_col='date'
)

join_df = join_df.drop(columns=['vaccinated'])

join_df.head()
```

```python
snap_df = join_df.groupby(['code'], as_index=False).last()

snap_df.head()
```

```python
# concat

snap_df = pd.merge(snap_df, vax_ddf, on=['code'])

snap_df.head()
```

```python
print('saving snapshotted df')
snap_df.to_csv(f'{data_path}/snap_parana.csv', index=False)
```
