```python
import pandas as pd

dataset_size = 'full'

all_data_path = './../data/join/'
data_path = f'./../data/join/{dataset_size}/'

rem_all_data_path = './../data/rem/'
rem_data_path = f'./../data/rem/{dataset_size}/'

vax_all_data_path = './../data/vax/'
vax_data_path = f'./../data/vax/{dataset_size}/'
```

```python
# read removed

rem_ddf = pd.read_csv(f'{rem_data_path}/removed_parana.csv',
    sep=',',
    dtype={
        'date': str,
        'code': float
    })

rem_ddf.head()
```

```python
# read vaccinated

vax_ddf = pd.read_csv(f'{vax_data_path}/vax_count_acc.csv',
    sep=',',
    dtype={
        'date': str,
        'code': str
    }
)

# vax_ddf = vax_ddf.astype({'code': 'float64'})
vax_ddf['code'] = pd.to_numeric(vax_ddf['code'], errors='coerce')

vax_ddf.head()
```

```python
# concat

ddf = pd.merge(vax_ddf, rem_ddf, on=['date', 'code'])

ddf.head()
```

```python
print('saving joined dfs')
ddf.to_csv(f'{data_path}/covid_parana.csv', index=False)
```
