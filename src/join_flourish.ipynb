```python
import pandas as pd
import numpy as np

dataset_size = 'full'

all_data_path = './../data/join/'
data_path = f'./../data/join/{dataset_size}/'
```

```python
# read joined

covid_ddf = pd.read_csv(f'{data_path}/covid_parana.csv',
    sep=',',
    dtype={})

covid_ddf.head()
```

```python
covid_ddf['week'] = pd.Int64Index(pd.DatetimeIndex(covid_ddf['date']).isocalendar().week)
covid_ddf['year'] = pd.DatetimeIndex(covid_ddf['date']).year

covid_ddf.head()
```

```python
df_grouped = covid_ddf.groupby('name')

df_year_groups = covid_ddf.groupby(['year', 'week'])

df_year_groups.head()
```

```python
flourish_df = covid_ddf[['name', 'year', 'week', 'deaths']]

flourish_df = pd.pivot_table(flourish_df,
    index=['name'],
    columns=['year', 'week'],
    aggfunc='last')

flourish_df[flourish_df.columns[0]].fillna(0, inplace=True)

flourish_df.head()
```

```python
print('saving flourish dfs')
flourish_df.to_csv(f'{data_path}/covid_deaths_parana_flourish.csv', index=True)
```
